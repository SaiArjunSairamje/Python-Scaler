### **Q1. Survival and age**
***

In the titanic dataset, to understand the impact of age and gender on the chances of survival, which of these option(s) would be a suitable choice?

* Boxplot can be used here. It shows that the median age of survivors in males is higher than in non-survivors.
* Boxplot can be used here. It shows that the median age of survivors in females is higher than in non-survivors.
* Barplot can be used here. It shows that the mean age of survivors is lesser than non-survivors in males.
* Histogram can be used here. It shows that the median age of survivors is higher than non-survivors in females.

```python
import seaborn as sns
titanic = sns.load_dataset('titanic')
```

**Correct Answers:**

* Boxplot can be used here. It shows that the median age of survivors in females is higher than in non-survivors.
* Bargraph can be used here. It shows that the mean age of survivors is lesser than non-survivors in males.

**Explanation:**

We can compare the ages of different sex on the basis of whether they survived or not, using boxplot:

```python
females = titanic[titanic['sex'] == 'female']
sns.boxplot(x="survived", y="age", data=females)
```

* After observing the boxplot above, we found that the median age of non-survivors in females is lesser than survivors.
* For comparing mean we can use barplot to compare mean using the same syntax as above but with an additional parameter **estimator = np. mean**

```python
sns.barplot(x='sex',y='age',data=titanic,palette='rainbow',hue='survived',estimator=np.mean)
```

* The mean age of survivors also appears to be a little younger than those who died in males.
* Histograms can't be used here as they can show visualization based on frequency but won't be able to compare the median according to survival and gender of passengers.
