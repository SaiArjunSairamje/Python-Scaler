**Given a NumPy array, your task is to multiple the even number by 2 and the odd number by 3**

This is where the **"Vectorization"** comes into picture where in we can **customize** the **element wise operatations** as per our needs...

**Example o/p of using "Vectorization"**

Capture.PNG



```python
#Given a NumPy array, your task is to multiple the even number by 2 and the odd number by 3
```

```python
def even_odd(x):
  if x%2 == 0:
    x = x*2
  else:
    x = x * 3

  return x
```

```python
even_odd(3) #using single integer
```

```python
even_odd(4)
```

```python
print(a)
```

```python
print(even_odd(a))  #this will throw an error as the def even_odd function is expcting a single integer value and not a list value
```

```python
func = np.vectorize(even_odd)  #to fix the above error we use "np.vectorize()"
```

# **Broadcasting**

**Broadcasting** is a powerful concept in **NumPy** that allows arrays with different shapes to be combined and operated together. This feature enables **NumPy** to handle **element-wise operations** on arrays of different sizes, which simplifies code and avoids the need for explicit loops.

**np.tile()** is a function in NumPy that allows you to create a new array by repeating an existing array (tile) along specified dimensions. This function is helpful when you want to replicate an array to create a larger array or to match the dimensions of another array for broadcasting.

The syntax for **np.tile()** is as follows:

> **`numpy.tile(A, reps)`**

Where:

* **A:** The input array to be tiled or repeated.
* **reps:** The number of times to repeat the input array along each dimension. It can be an integer or a tuple of integers.

**Example o/p of "np.tile()"**

Capture.PNG

```python
b = np.tile(a, (3,1))   #this gives replicated data of the array "a" --> 3 rows & 1 column o/p replcations of 1D array(i.e. "a" variable)
print(b)
```

```python
b = np.tile(a, (3,2))  #this will give a single row but 2 copies in that single row    #NOTE: (3, 2) is not a shape. Its 3 rows with 2 copies on each rows
print(b)
```
