### **How to find duplicate rows in Pandas ???**

The **df.duplicated()** method in Pandas is used to identify duplicate rows in a DataFrame. It returns a Boolean Series that indicates whether each row is a duplicate or not.

The resulting Boolean Series indicates **False** for non-duplicate rows and **True** for the duplicate row(s).


```python
df.duplicated()
```

### **How to extract duplicate rows ???**

> **`Approach (1):`** Normal `indexing` method with **`duplicated()`** function

> **`Approach (2):`** Using the **`loc`** indexer with **`duplicated()`** function

To extract the duplicate rows using the **loc** indexer, we use **df.loc[duplicates]**, where **duplicates** is the boolean Series that indicates which rows are duplicates. The resulting DataFrame **duplicate_rows** contains only the rows that are duplicates.



```python
# Approach (1)

df[df.duplicated()]
```

```python
# Approach (2)

df.loc[df.duplicated()]
```

**Why we can't extract duplicated rows using iloc() ???**

The **iloc()** function in Pandas is primarily used for integer-based indexing and slicing of rows and columns in a DataFrame. It allows you to access specific rows and columns based on their integer positions.

While **iloc()** can be used to select rows based on their positions, it does not have a direct functionality to identify and extract duplicated rows. This is because duplicate rows are not necessarily consecutive in the DataFrame, and their positions may vary based on the sorting or filtering criteria applied to the data.

```python
df.iloc[df.duplicated()]  # this will throw an error message
```

### **How to remove the duplicate rows ???**

`Note` that **df.loc[df.duplicated()]** returns a new DataFrame that includes only the duplicate rows. If you want to retain the original DataFrame and remove the duplicate rows, you can use **df.drop_duplicates()** as shown in the previous response.

```python
df.tail()
```

```python
df.drop_duplicates()
```

```python
df.tail()  # the original dataFrame still has the duplicate row and we need to fix it...
```

**Since there are 2 rows having same values, Which duplicate row to be removed ???**

The **drop_duplicates()** method in Pandas is used to remove duplicate rows from a DataFrame. It provides the **keep** parameter to control which duplicate values to keep. There are 2 primary ways to use **drop_duplicates()** with different values for the **keep** parameter:

> **`Approach (1):` keep='first'** (default)

> **`Approach (2):` keep='last'**

**Approach (1): keep='first'** (default)         
By default, the **drop_duplicates()** method keeps the first occurrence of each duplicated row and removes the subsequent occurrences.

```python
df.drop_duplicates(keep = "first")
```

**Approach (2): keep='last'**            
Setting **keep='last'** as the parameter value for **drop_duplicates()** keeps the last occurrence of each duplicated row and removes the previous occurrences.

```python
df.drop_duplicates(keep = "last")
```

          
**pd.join()** (**`NOTE:`** Additional topic external from **Scaler**)

This function is used to join two DataFrames based on their indices. It is similar to the **merge()** operation but uses the index values of DataFrames for the join operation.

In this example, **df1.join(df2)** performs a join operation on the indices of **df1** and **df2**, resulting in a new DataFrame **joined_df**. The columns from both DataFrames are combined based on the matching indices.



```python
# Create two DataFrames with different column names
df1 = pd.DataFrame({'A': [1, 2, 3],
                    'B': [4, 5, 6]},
                   index=['X', 'Y', 'Z'])
df2 = pd.DataFrame({'C': [7, 8, 9],
                    'D': [10, 11, 12]},
                   index=['X', 'Y', 'Z'])

# Join DataFrames based on index
joined_df = df1.join(df2)

# Print the joined DataFrame
print(joined_df)# Join DataFrames based on index
```

**Give me only those columns where the vote_average is greater than 7 and the year is on or after 2015**

The code **data.loc[(data["vote_average"] > 7) & (data["year"] >= 2015)]** demonstrates the usage of two operators, namely the comparison operator **>** and the logical operator **&**.

Here are explanations of these operators and additional commonly used operators in pandas:

**(1) Comparison Operators:**

> **>**: It compares if the value on the left is greater than the value on the right.  
> **<**: Less than operator. It compares if the value on the left is less than the value on the right.      
> **>=**: It compares if the value on the left is greater than or equal to the value on the right.           
> **<=**: It compares if the value on the left is less than or equal to the value on the right.            
> **==**: It compares if the value on the left is equal to the value on the right.   
> **!=**: It compares if the value on the left is not equal to the value on the right.

**(2) Logical Operators:**

> **&**: Element-wise logical AND operator. It performs element-wise logical AND between two conditions, resulting in True if both conditions are True, and False otherwise.  
> **|**: Element-wise logical OR operator. It performs element-wise logical OR between two conditions, resulting in True if at least one condition is True, and False otherwise.      
> **~**: Element-wise logical NOT operator. It negates a condition, resulting in True if the condition is False, and False if the condition is True.  

These operators can be used within the indexing or selection operations of pandas, such as **loc[]** or **iloc[]**, to filter or select specific rows based on conditions.


```python
data.loc[ (data["vote_average"] > 7) & (data["year"] >= 2015) ]
```
