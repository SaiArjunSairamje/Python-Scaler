**How to find duplicate rows in Pandas ???**

The **df.duplicated()** method in Pandas is used to identify duplicate rows in a DataFrame. It returns a Boolean Series that indicates whether each row is a duplicate or not.

The resulting Boolean Series indicates **False** for non-duplicate rows and **True** for the duplicate row(s).


```python
df.duplicated()
```

**How to extract duplicate rows ???**

To extract the duplicate rows using the **loc** indexer, we use **df.loc[duplicates]**, where **duplicates** is the boolean Series that indicates which rows are duplicates. The resulting DataFrame **duplicate_rows** contains only the rows that are duplicates.



```python
df.loc[df.duplicated()]
```

Note that **df.loc[duplicates]** returns a new DataFrame that includes only the duplicate rows. If you want to retain the original DataFrame and remove the duplicate rows, you can use **df.drop_duplicates()** as shown in the previous response.

```python
df.drop_duplicates()
```

```python
df.tail()  # the original dataFrame still has the duplicate row and we need to fix it...
```

**Which duplicate row to be removed ???**

The **drop_duplicates()** method in Pandas is used to remove duplicate rows from a DataFrame. It provides the **keep** parameter to control which duplicate values to keep. There are two primary ways to use **drop_duplicates()** with different values for the **keep** parameter:

> **Approach (1): keep='first'** (default)

By default, the **drop_duplicates()** method keeps the first occurrence of each duplicated row and removes the subsequent occurrences.

```python
df.drop_duplicates(keep = "first")
```

> **Approach (2): keep='last'**

Setting **keep='last'** as the parameter value for **drop_duplicates()** keeps the last occurrence of each duplicated row and removes the previous occurrences.

```python
df.drop_duplicates(keep = "last")
```

> **(3) Data Combination or Data Merging Operations:**

*   **pd.concat()**
*   **pd.merge()**
*   **pd.join()**

**pd.concat()**:        

This function is used to concatenate two or more DataFrames along a particular axis. It can be used to combine DataFrames vertically (along rows) or horizontally (along columns).

```python
users = pd.DataFrame({"userid":[1, 2, 3], "name":["AJ", "Manoj", "Mukesh"]})
users
```

```python
msgs = pd.DataFrame({"userid":[1, 1, 2, 4], "msg":['hmm', "acha", "theek hai", "nice"]})
msgs
```

```python
pd.concat([users, msgs], axis = 0) # Vertical Stack
```

```python
pd.concat([users, msgs], axis = 0, ignore_index = True)
```

```python
pd.concat([users, msgs], axis = 1) # Horizonal Stack
# NOTE: When ever there is a "NaN" value in a numerical column, all values will become "Float" datatype
```

**pd.merge()**:    
      
This function is used to merge two DataFrames based on common columns or indices. It performs database-style joins (e.g., inner join, outer join, etc.) to combine DataFrames.

```python
users
```

```python
msgs
```

```python
users.merge(msgs, on = "userid") # Inner Join
```

```python
# How to perform particular type of Join ???

users.merge(msgs, on = "userid" , how = "outer") # Outer Join
```

```python
users.merge(msgs, on = "userid", how = "left") # Left Join
```

```python
users.merge(msgs, on ="userid", how = "right") # Right Join
```

**What to do when the column names of the 2 DataFrames are not the same ???**    
**How can we merge DataFrames in such a case ???**

By explicitly specifying the columns to merge on using **left_on** and **right_on** parameters, you can handle scenarios where the column names in the DataFrames are not the same but there is a common column or index to merge on.






```python
users.rename(columns = {"userid" : "id"} , inplace = True)
```

```python
users
```

```python
msgs
```

```python
users.merge(msgs, left_on = "id", right_on = "userid")
```

          
**pd.join()** (**`NOTE:`** Additional topic external from Scaler)

This function is used to join two DataFrames based on their indices. It is similar to the **merge()** operation but uses the index values of DataFrames for the join operation.

In this example, **df1.join(df2)** performs a join operation on the indices of **df1** and **df2**, resulting in a new DataFrame **joined_df**. The columns from both DataFrames are combined based on the matching indices.



```python
# Create two DataFrames with different column names
df1 = pd.DataFrame({'A': [1, 2, 3],
                    'B': [4, 5, 6]},
                   index=['X', 'Y', 'Z'])
df2 = pd.DataFrame({'C': [7, 8, 9],
                    'D': [10, 11, 12]},
                   index=['X', 'Y', 'Z'])

# Join DataFrames based on index
joined_df = df1.join(df2)

# Print the joined DataFrame
print(joined_df)
```

The code **movies["director_id"].isin(directors["id"])** checks if the values in the **"director_id"** column of the DataFrame **movies** are present in the **"id"** column of the DataFrame **directors**. It returns a boolean Series with **True** for the rows where the condition is met and **False** otherwise.

In this example, the boolean Series **is_valid_director** is generated by comparing the "director_id" column of **movies** with the "id" column of **directors**. It indicates which rows in **movies** have a valid director ID present in **directors**. The resulting Series contains **True** for the rows where the condition is satisfied, and **False** otherwise.

The code **np.all(movies["director_id"].isin(directors["id"]))** checks if all the values in the "director_id" column of the DataFrame **movies** are present in the "id" column of the DataFrame **directors**. It returns **True** if all the values in the "director_id" column are found in the "id" column, and **False** otherwise.

In this example, **np.all(movies["director_id"].isin(directors["id"]))** checks if all the values in the "director_id" column of **movies** are present in the "id" column of **directors**. Since the result is **True**, it means that all the director IDs in **movies** are found in **directors**.
