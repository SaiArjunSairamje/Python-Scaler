Matplotlib's anatomy consists of the following key components:

* **Figure**: The top-level container representing the entire plot.    
* **Axes**: The individual plot within a figure, including coordinates, labels, and ticks.    
* **Subplots**: Grid of axes allowing multiple plots in a single figure.     
* **Axis**: Represents the x-axis or y-axis of a plot, including ticks and labels.   
* **Artist**: Fundamental building blocks representing graphical elements like lines, patches, and texts.

### **Bar Chat - "Matplotlib" & "Seaborn"**

> **`Bar Chart with Matplotlib:`**        
**"Matplotlib"** is a popular plotting library in Python that can be used to create **bar charts**. You can use the **plt.bar()** function to generate a **bar chart**.

> **`Bar Chart with Seaborn:`**           
**"Seaborn"** is another Python library **built on top of "Matplotlib"** that provides a higher-level interface for creating aesthetic and informative visualizations. To create a **bar chart** with **"Seaborn"**, you can use the **sns.barplot()** function.


```python
x_bar = cat_counts.index
print(x_bar)
```

```python
y_bar = cat_counts
```

```python
# Bar Chart using "Matplotlib"

plt.bar(x_bar, y_bar)
```

```python
plt.figure(figsize = (12, 8))
plt.bar(x_bar, y_bar)
```

```python
plt.bar(x_bar, y_bar)
plt.xticks(rotation = 90, fontsize = 12)
```

```python
plt.bar(x_bar, y_bar, width = 0.2)
plt.xticks(rotation = 90, fontsize = 12)
```

```python
# plt.figure(figsize = (8, 5))
plt.bar(x_bar, y_bar, color = 'orange')
plt.title("Games per Genre", fontsize = 15)
plt.xlabel("Genre", fontsize = 12)
plt.ylabel("Count" , fontsize = 12)
plt.xticks(rotation = 90, fontsize = 15)
plt.yticks(fontsize = 12)
```

**Bar Chart using "Matplotlib"**       
**`Link:`** https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html

```python
plt.figure(figsize = (12, 5))       #Set figure size
plt.bar(x_bar, y_bar, color = 'orange')     #Create bar chart
plt.title("Games per Genre", fontsize = 15)     #Set chart title
plt.xlabel("Genre", fontsize = 12)      #Set x-axis label
plt.ylabel("Count" , fontsize = 12)     #Set y-axis label
plt.xticks(rotation = 90, fontsize = 15)      #Rotate x-axis tick labels
plt.yticks(fontsize = 12)       #Set y-axis tick label font size
plt.show()        #Display the chart
```

### **KDE - "Matplotlib" & "Seaborn"**

**KDE (Kernel Density Estimation)** is a technique used to estimate the probability density function of a continuous random variable. It provides a smoothed representation of the underlying probability distribution.

In terms of visualizations, both **Matplotlib** and **Seaborn** offer methods to create **KDE plots**.

> **`KDE Plot with Matplotlib:`**      
**Matplotlib** provides the **plt.plot()** or **plt.fill_between()** functions to create a **KDE plot**.

> **`KDE Plot with Seaborn:`**        
**Seaborn** offers the **sns.kdeplot()** function to create a **KDE plot**.

```python
#KDE - probability distribution

sns.kdeplot(data["Year"])
```

### **Box Plot - "Matplotlib" & "Seaborn"**

A **box plot**, also known as a **box-and-whisker plot**, is a graphical representation of the distribution of a dataset. It displays the summary statistics such as median, quartiles, and outliers. Both **Matplotlib** and **Seaborn** provide methods to create box plots.

> **`Box Plot with Matplotlib:`**       
In **Matplotlib**, you can use the **plt.boxplot()** function to create a **box plot**.

> **`Box Plot with Seaborn:`**       
**Seaborn** offers the **sns.boxplot()** function to create a **box plot**.

```python
plt.figure(figsize = (10,7))
sns.boxplot(y = data["Global_Sales"])
plt.yticks(fontsize = 20)
plt.ylabel('Global Sale')
plt.title("Boxplot Intro")
plt.show()
```
