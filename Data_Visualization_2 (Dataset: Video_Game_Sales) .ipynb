```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("vgsales.csv")
```

```python
data.head()
```

```python
# Which name of the game is present how many time in this data???
data["Name"].value_counts()
```

**NOTE: Till now we had worked with Univariate Analysis. Today we will start with Bivariate Analysis (i.e. 2 columns we will be using for plotting the graph)...**

```python
# What is the year on year trends of "Ice Hockey" ???
ih = data.loc[data["Name"] == "Ice Hockey"]
```

```python
ih.head()
```

```python
#N-N (i.e. Numerical-Numerical attributes)
```

```python
# Yearly trend of North America's sales for Ice Hockey game...
plt.title("NA Sales for Ice Hockey")
sns.lineplot(x = "Year", y= "NA_Sales", data = ih)
plt.xlabel("Year")
plt.ylabel("NA Sales")
plt.show()
```

```python
# How to change the color of the draft???
plt.title("NA Sales for Ice Hockey")
sns.lineplot(x = "Year", y= "NA_Sales", data = ih, color = 'r')
plt.xlabel("Year")
plt.ylabel("NA Sales")
plt.show()
```

```python
plt.title("NA Sales for Ice Hockey")
sns.lineplot(x = "Year", y= "NA_Sales", data = ih, color = 'k')   #color = black
plt.xlabel("Year")
plt.ylabel("NA Sales")
plt.show()
```

```python
# How to zoom/limit into specific segment of the plot as per our needs???
plt.title("NA Sales for Ice Hockey")
sns.lineplot(x = "Year", y= "NA_Sales", data = ih, color = 'k')
plt.xlim(left = 1995, right = 2010)   #Zooming/limiting the x-axis segment of the flot...
plt.xlabel("Year")
plt.ylabel("NA Sales")
plt.show()
```

The double line graph appearing above seem to be confusing as we are unable to find which line is for which attribute.

Hence, to fix that we have 2 approaches...

```python
#version1
sns.lineplot(x = "Year", y= "NA_Sales", data = ih, label = "Ice Hockey")
sns.lineplot(x = "Year", y= "NA_Sales", data = baseball, label = "Baseball")
```

```python
#version2 (NOTE: preferred approach)
sns.lineplot(x = "Year", y= "NA_Sales", data = ih)
sns.lineplot(x = "Year", y= "NA_Sales", data = baseball)
plt.legend(["Ice Hockey", "Baseball"])
plt.show()
```

```python
#version2
sns.lineplot(x = "Year", y= "NA_Sales", data = ih)
sns.lineplot(x = "Year", y= "NA_Sales", data = baseball)
plt.legend(["Ice Hockey", "Baseball"], loc = (-0.5, 0.5))
plt.show()
```

How to change the location of the **Legend** as per our needs???

There are 2 approaches to get this fixed...

```python
# Approach 1
sns.lineplot(x = "Year", y= "NA_Sales", data = ih)
sns.lineplot(x = "Year", y= "NA_Sales", data = baseball)
plt.legend(["Ice Hockey", "Baseball"], loc = (-0.5, 0.5))
plt.show()
```

```python
# What to do if I don't want the Legend appear as a box as shown previously???
sns.lineplot(x = "Year", y= "NA_Sales", data = ih)
sns.lineplot(x = "Year", y= "NA_Sales", data = baseball)
plt.legend(["Ice Hockey", "Baseball"], loc = (-0.5, 0.5), frameon = False)
plt.show()
```

```python
# Approach 2 (i.e. Standard approach without having to mention any specific location details)
sns.lineplot(x = "Year", y= "NA_Sales", data = ih)
sns.lineplot(x = "Year", y= "NA_Sales", data = baseball)
plt.legend(["Ice Hockey", "Baseball"], loc = 'lower right', frameon = False)
plt.show()
```

```python
sns.lineplot(x = "Year", y= "NA_Sales", data = ih)
sns.lineplot(x = "Year", y= "NA_Sales", data = baseball)
plt.legend(["Ice Hockey", "Baseball"], loc = 'lower left', frameon = False)
plt.show()
```

```python
# (IMP - very few people know this)
# So far we have the Legend having 2 rows & 1 column view on the o/p.
# What to do if we want one beside another (i.e. 1 row & 2 columns for Legend) ???
sns.lineplot(x = "Year", y= "NA_Sales", data = ih)
sns.lineplot(x = "Year", y= "NA_Sales", data = baseball)
plt.legend(["Ice Hockey", "Baseball"], loc = 'lower left', ncol= 2, frameon = False)
plt.show()
```

```python
# What is the maximum North American sales for Ice Hockey ???
print(max(ih["NA_Sales"]))    #the above plot also highlights the same...
```

```python
# How can be add test for my graph having the x-coordinate and y-coordinate that has the maximum sales...
sns.lineplot(x = "Year", y= "NA_Sales", data = ih)
sns.lineplot(x = "Year", y= "NA_Sales", data = baseball)
plt.legend(["Ice Hockey", "Baseball"], loc = 'lower left', ncol= 2, frameon = False)
plt.text(2002, max(ih["NA_Sales"]), "Max Ice Hockey NA Sales")   # this was done using hard coding approach but can do generic as well...
plt.show()
```
