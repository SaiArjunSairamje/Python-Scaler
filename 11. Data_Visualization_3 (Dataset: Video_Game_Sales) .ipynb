#### **(2) Catplot - "Seaborn" using "Color" of o/p image (`NOTE:` outside Scaler)**
***

For **CCN (Categorical-Categorical-Numerical)** analysis, where you have **two categorical variables** and **one numerical variable**, you can use **sns.catplot** in **Seaborn** to create a more structured and versatile plot. By using **sns.catplot**, you can easily facet the data by the third categorical variable.

In this code, **kind='box'** inside **sns.catplot** specifies that you want a box plot for the numerical variable (**Global_Sales**) and **hue='Genre'** differentiates the data points based on the 'Genre' category. The **height** and **aspect** parameters control the size and aspect ratio of the plot to make it visually appealing.

**sns.catplot** creates a FacetGrid, which allows you to create multiple plots based on the values of a categorical variable (**Genre** in this case). This way, you can easily visualize and compare the distribution of 'Global_Sales' for each 'Publisher' across different 'Genre' categories.

**Link:** https://seaborn.pydata.org/generated/seaborn.catplot.html

```python
# Another plot (Outside Scaler)
plt.figure(figsize=(15, 10))
sns.catplot(x='Publisher', y='Global_Sales', hue='Genre', kind='box', data=top3_data, height=6, aspect=2)
plt.xlabel('Genre', fontsize=12)
plt.ylabel('Global Sales', fontsize=12)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.title('Global Sales for each Publisher', fontsize=15)
plt.show()
```

```python
'''
# Personal Notes

Pairplot:
sns.pairplot is a Seaborn function that creates a grid of scatterplots, allowing you to visualize the pairwise relationships between
multiple numerical variables in your dataset. It displays the scatterplot of each numerical variable against every other numerical variable, along with
histograms on the diagonal for each variable.

Correlation Matrix:
A correlation matrix is a tabular representation of the correlations between multiple numerical variables in your dataset. It shows how strongly
 two variables are related to each other. The values range from -1 (perfect negative correlation) to 1 (perfect positive correlation), with
 0 indicating no correlation.

Heatmap:
A heatmap is a graphical representation of data using colors in a two-dimensional matrix. In the context of a correlation matrix, the heatmap
visually represents the correlation coefficients between numerical variables. It uses colors to show the strength and direction of the correlation
between variables, making it easier to identify patterns and relationships.
'''
```

#### **Correlation & Causation**

**Correlation** and **causation** are two fundamental concepts in statistics and data analysis. They describe the relationship between two variables and help us understand the nature of their association. However, it is essential to understand that **correlation** does **not** imply **causation**.



```python
# Correlation
'''
Correlation: -1 to 1
corr(a,b) --> 1 (Strongly Positive Correlation) --> Ex: NA_Sales Vs Global_Sales are +ve correlated
corr(a,b) --> -1 (Strongly Negative Correlation) --> Ex: Rank Vs Global_sales are -ve correlated
corr(a,b) --> 0 (No Correlation)
'''

# corr(a,a) -> +1  --> Correlation b/w same variable will have "Strongly Positive Correlation'

# NOTE: There are 2 famous terms in Statistic named "Correlation" & "Causation"
# Ex: During summer, the no. of sharks in the sea incrases & no. of ice creams also increases --> diff b/w "Correlation" & "Causation" takes place here..
```

#### **(2) HeatMap**
***

```python
# HeatMap
# Using the Correlation Matrix we are able to apply the HeatMap to obtain a better o/p data...
sns.heatmap(top3_data.corr())
```

Just looking at the image above we are able to see what variables are **Highly Postive Correlated**, what is **Highly Negative Correlated** and what are **Not Correlated** at all...

But we can still see that we are **missing the numerical value** for a better visualization. Hence, let's merge both the **Colors & Numerical Values** to get best o/p in the next code...

```python
# combination of Color & Numbers --> "lighter the color, higher the correlation" & "darker the color, lesser the correlation"
# When annot=True, the heatmap cells will be annotated with the numerical values of the correlation coefficients, which are computed from the correlation matrix
sns.heatmap(top3_data.corr(), annot = True)
```

```python
# How to make the o/p image more beautiful ???
# We can use "cmap" feature as "Blues" that has all the Numerical Values under one color
sns.heatmap(top3_data.corr(), annot = True, cmap = "Blues")   # "darker the color, lesser the correlation" & "lesser the color, higher the correlation"
```
